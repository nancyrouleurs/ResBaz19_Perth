```python

import numpy as np 
import pandas as pd 
from sklearn import datasets, cluster, model_selection
import matplotlib.pyplot as plt
```

```python
data=datasets.make_blobs(n_samples=100,n_features=2,cluster_std=2,center_box=(-10.0, 10.0),centers=4,random_state=6)
plt.scatter(data[0][:,0],data[0][:,1])
```

```python
x=data[0]
vals=[]
nrange=range(1,10)
for n in nrange:
    res=cluster.k_means(x,n_clusters=n)
    vals.append(res[2])
```

```python
plt.plot(nrange,vals)
```

```python
n=4
res=cluster.k_means(x,n_clusters=n)

for i in range(n):
    idx= res[1]==i
    plt.scatter(x[idx,0],x[idx,1])
    plt.scatter(res[0][i,0],res[0][i,1],marker='x',c='k')
```

**Now let's see how it performs on higher dimensions**

```python
data=datasets.make_blobs(n_samples=100,n_features=8,cluster_std=2,center_box=(-10.0, 10.0),centers=6,random_state=11)
vals=[]
nrange=range(1,10)
for n in nrange:
    res=cluster.k_means(data[0],n_clusters=n)
    vals.append(res[2])
plt.plot(nrange,vals)
```
