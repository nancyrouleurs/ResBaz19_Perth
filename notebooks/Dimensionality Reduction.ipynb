```python
import numpy as np
from sklearn import datasets, decomposition
import matplotlib.pyplot as plt
import sklearn
```

```python
data=datasets.load_digits()
```

```python
print(data.DESCR)
```

```python
x=data.data
y=data.target
```

```python
ngrid=6
plt.figure(figsize=(16,16))
for i in range(ngrid**2):
    plt.subplot(ngrid,ngrid,i+1)
    idx=np.random.randint(len(x))
    digit=x[idx,:]
    digit=digit.reshape((8,8))
    plt.imshow(digit,'gray')
    plt.title(y[idx])
        
```

```python
x[0]
```

```python
# we want to represent 64 dimensions using only 2
pca=decomposition.PCA(n_components=2)
```

```python

x_2d=pca.fit_transform(x)
```

```python
plt.figure(figsize=(6,6))
for i in range(10):
    idx= y==i
    plt.scatter(x_2d[idx,0],x_2d[idx,1])
plt.legend(range(10))
```
