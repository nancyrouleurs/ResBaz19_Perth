```python

import numpy as np 
import pandas as pd 
from sklearn import datasets, ensemble, model_selection
import matplotlib.pyplot as plt
import pandas_profiling as pp
import seaborn as sns
import xgboost as xgb
```

```python
data=datasets.load_boston()
```

```python
print(data.DESCR)
```

```python
df=pd.DataFrame(data.data,columns=data.feature_names)
df['Target']=data.target
```

```python
df.describe()
```

```python
pp.ProfileReport(df)
```

```python
sns.pairplot(df)
```

```python
x,y=data.data,data.target
```

```python
xtr,xval,ytr,yval=model_selection.train_test_split(x,y, test_size=0.3, random_state=23)
```

```python
model=xgb.XGBRegressor(random_state=25,objective='reg:squarederror')
```

```python
model.fit(xtr,ytr)
```

```python
model.score(xtr,ytr),model.score(xval,yval)
```

That's pretty good. Let's see if we can make this better by hyperparameer optimization.

```python
params={'learning_rate':[0.05,0.1,.2],'n_estimators':[100,200,400],'max_depth':[2,3],'subsample':[.4,.6,.8]}

grid_search=model_selection.GridSearchCV(xgb.XGBRegressor(random_state=7,objective='reg:squarederror'),
                                         param_grid=params,
                                         cv=5)
```

```python
%time grid_search.fit(xtr,ytr)
```

```python
for i,param in enumerate(grid_search.cv_results_['params']):
    print(param,grid_search.cv_results_['mean_test_score'][i])
```

```python
grid_search.best_params_
```

```python
model=grid_search.best_estimator_
model.score(xtr,ytr),model.score(xval,yval)
```
